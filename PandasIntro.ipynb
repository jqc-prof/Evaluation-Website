```python
import os

#create a directory joining the parent directory ".." to "data" and check to see
#if the directory already exists
os.makedirs(os.path.join('..', 'data'), exist_ok=True)

#create file under directory named house_tiny.csv while joining the filepath
data_file = os.path.join('..', 'data', 'house_tiny.csv')

#write the data file
with open(data_file, 'w') as f:
  #first row is column labels
  f.write('''NumRooms,RoofType,Price
  NA,NA,127500
  2,NA,106000
  4,Slate,178100
  NA,NA,140000''')
```

```python
import pandas as pd
data = pd.read_csv(data_file, skipinitialspace=True)

#remove NAN values completely
#select columns via integer-location based indexing (iloc) to separate out
#columns corresponding to input versus target values
inputs, targets = data.iloc[:, 0:2], data.iloc[:, 2]
inputs = pd.get_dummies(inputs, dummy_na=True)

#replace all nan entries with mean value of corresponding column
inputs = inputs.fillna(inputs.mean())
print(inputs)
```

```python
import torch
#load values into a tensor
X = torch.tensor(inputs.to_numpy(dtype=float))
Y = torch.tensor(targets.to_numpy(dtype=float))
X,Y
```
